# Creating Vector and matrix for M.L Style Data



```python
x = np.array(
    [
        [2,30],# Sample 1
        [5,80],# Sample 2
        [1,20]  #Sample 3 
    ]
)
y = np.array([0,1,0])
print("Input Data matrix X :\n" ,x)
print("X shape: " , x.shape)
                  #(n_sample ,n_features) 
print("Output vector y:\n" , y)
print("y shape :", y.shape)
                  # (n_samples)
# A single sample as a vector 
x1 = x[0]  # first row 
print("\n First sample as a vector : "  , x1)
print("x1 shape : " , x1.shape ) 
```

# Basic Operation : Addition ,Scaler , Multiplication , transpose 

Two Vectors with same length

```python
v1 = np.array([1,2,3]) 
v2 = np.array([4,5,6])
print("v1:",v1)
print("v2:",v2)
```

Vector addition 

```python
v_add = v1+v2
print("\n Vector addition v1 + v2:" , v_add)
```

Scaler Multiplication

```python
v1_scaled = 2*v1
print("Scaler multiplication 2*v1:" ,v1_scaled)
```

# Matrix Multiplication 

```python
a = np.array(
    [
        [1,2],
        [3,4]
    ])
b = np.array(
    [
        [10,20],
        [30,40]
    ])
```

```python
print("\n Matrix A:\n",a)
print("Matrix B:\n" , b)
```

Matrix Addition

```python
a_add_b = a+b
print("\n matrix addition a+b:\n", a_add_b)
```

Scaler Multiplication

```python
a_scaled = 0.5*a
print("Scaler multiplication 0.5*A:\n" , a_scaled)
```

Transpose

```python
a_t = a.T 
print("\n Transpose of A:\n" ,a_t)
print("a shape:" , a.shape," (a.t shape :", a_t.shape)
```
